```python
###### Fonction de prÃ©diction pour dash

# moy des prix des valeurs foncieres des locaux en fonction des departements en 2021
moy_dep_l_2021 = pd.read_csv(filepath_or_buffer="C:/Users/pierr/Documents/1_projet_python/Valeur_fonciere/Depot_data/moy_departement_locaux_2021.csv")
# moy des prix des valeurs foncieres des dÃ©pendances en fonction des departements en 2021
moy_dep_d_2021 = pd.read_csv(filepath_or_buffer="C:/Users/pierr/Documents/1_projet_python/Valeur_fonciere/Depot_data/moy_departement_dependance_2021.csv")

prix_m2_com_region = pd.read_csv(filepath_or_buffer="C:/Users/pierr/Documents/1_projet_python/Valeur_fonciere/Depot_data/prix_m2_com_region.csv")

#On importe le modÃ¨le pickle pour le prix
prix_model_maison = pickle.load(open('reg_prix_maison.pkl', 'rb'))
prix_model_appart = pickle.load(open('reg_prix_appart.pkl', 'rb'))



def prediction(Type_loc,Surface,region, commune):

    prix_m2 = prix_m2_com_region[(prix_m2_com_region['Commune'] == commune) & (prix_m2_com_region['Code departement'] == region)]['Prix m2 moyen commune']
    d = {'Surface reelle bati' : Surface, 'Prix m2 moyen commune' : prix_m2}
    df = pd.DataFrame(data = d)
    if(Type_loc == 'Maison'):
        result = pd.DataFrame(prix_model_maison.predict(df)).iloc[0]
    elif (Type_loc == 'Appartement'):
        result = pd.DataFrame(prix_model_appart.predict(df)).iloc[0]
    elif (Type_loc == 'DÃ©pendance'):
        result = moy_dep_d_2021[moy_dep_d_2021['Code departement']== region]['Valeur fonciere']
    else :
        result = moy_dep_l_2021[moy_dep_d_2021['Code departement']== region]['Valeur fonciere'] + 10000

    return result.iloc[0]
```

### Test de notre modÃ¨le sur 2018, 2019 ,2020 ,2021 et stat descriptive pour vÃ©rifier 

```python
def test_meansquare(df):

    df_m = df[df['Type local'] == 'Maison'][['Surface reelle bati', 'Prix m2 moyen commune']]
    df_a = df[df['Type local'] == 'Appartement'][['Surface reelle bati', 'Prix m2 moyen commune']]
    df_l = df[df['Type local'] == 'Local industriel. commercial ou assimilÃ©']
    #print(df_l)
    df_d = df[df['Type local'] == 'DÃ©pendance']


    df.loc[df_m.index,'y_pred'] = prix_model_maison.predict(df_m)
    df.loc[df_a.index,'y_pred'] = prix_model_appart.predict(df_a)
    df.loc[df_l.index,'y_pred'] = df_l.apply(return_valuerfm, axis=1)
    df.loc[df_d.index,'y_pred'] = df_d.apply(return_valeurfdep, axis=1)

    return df
```

```python
##### Test de la fonction sur 2021
url2 = "C:/Users/pierr/Documents/1_projet_python/Valeur_fonciere/Depot_data/concat_prix_m2_lati.csv"
df = pd.read_csv(filepath_or_buffer= url2 ,sep='|')

df_2021 = df[df['year']==2021]
df_2021['Code departement'] = df_2021['Code departement'].astype(str)
df_2020 = df[df['year']==2020]
df_2020['Code departement'] = df_2020['Code departement'].astype(str)
df_2019 = df[df['year']==2019]
df_2019['Code departement'] = df_2019['Code departement'].astype(str)
df_2018 = df[df['year']==2018]
df_2018['Code departement'] = df_2018['Code departement'].astype(str)
```

```python
df_2021_rt = test_meansquare(df_2021)
df_2020_rt = test_meansquare(df_2020)
df_2019_rt = test_meansquare(df_2019)
df_2018_rt = test_meansquare(df_2018)
```

```python
from sklearn.metrics import mean_squared_error
mse_2021 = mean_squared_error(y_pred=df_2021_rt['y_pred'],y_true=df_2021_rt['Valeur fonciere'], squared=False)
mse_2020 = mean_squared_error(y_pred=df_2020_rt['y_pred'],y_true=df_2020_rt['Valeur fonciere'], squared=False)
mse_2019 = mean_squared_error(y_pred=df_2019_rt['y_pred'],y_true=df_2019_rt['Valeur fonciere'], squared=False)
mse_2018 = mean_squared_error(y_pred=df_2018_rt['y_pred'],y_true=df_2018_rt['Valeur fonciere'], squared=False)
```
