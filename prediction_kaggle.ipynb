```python
d_type = {
    'Commune' : 'str',
    'Code commune' : 'str',
    'Code departement' : 'str',
    'Code postal' : 'str',
    'Surface reelle bati' : 'float',
    'Nombre pieces principales' : 'float',
    'Surface terrain' : 'float'
}

df = df_kaggle.astype(d_type)
df['Date mutation'] = pd.to_datetime(df['Date mutation'], format="%d/%m/%Y")

df['month'] = df['Date mutation'].dt.month
df['year'] = df['Date mutation'].dt.year



#Garder les colonnes pertinentes
df
```

# latitude longitude

```python
#Import des données
lat = pd.read_csv(filepath_or_buffer= path + "communes-departement-region.csv", sep=",", decimal=".")
#On vire les valeurs qui ne nous interesse pas
lat = lat[['code_postal','latitude', 'longitude','code_commune','code_departement']]
#Suppr valeur vide
lat = lat.dropna(axis=0)

print(lat.shape) #(38932, 5)

#Transtypage
d_type = {
    'code_postal':'float',
    'latitude':'float',
    'longitude':'float',
    'code_commune':'float',
    'code_departement':'str'
}


lat = lat.astype(d_type)
lat['code_commune'] = lat['code_commune'].astype(int).astype(str)
lat['code_postal'] = lat['code_postal'].astype(str)

#Changement des noms
lat = lat.rename(columns={'code_departement': 'Code departement', 'code_commune': 'Code commune', 'code_postal' : 'Code postal'})

lat.drop_duplicates(inplace = True)
print(lat.shape) #(36216, 5)

#Fusion
#Union des deux tables
df = pd.merge(left=df, right=lat, on=['Code departement','Code commune', 'Code postal'], how='left')

print(df.shape) #(378041, 38)
```

# Remplacement valeurs vides

```python
df_fin = df_test[['Type local','Surface reelle bati','Nombre pieces principales','year','latitude','longitude','Prix m2 moyen region','Prix m2 moyen commune']]
```

```python
df_fin.shape
```

```python
df_fin.isna().sum()
```

```python
#Remplacement des vlauers vides
df_fin['Surface reelle bati'] = df_fin['Surface reelle bati'].fillna(value = df_fin['Surface reelle bati'].median())
df_fin['Nombre pieces principales'] = df_fin['Nombre pieces principales'].fillna(value = 4)
#On remplace les valeurs nulles lat et longi par zero
values = {"latitude": 0, "longitude":0}
df_fin = df_fin.fillna(value = values)

df_fin['Prix m2 moyen commune'].fillna(df_fin['Prix m2 moyen region'], inplace= True)
```
